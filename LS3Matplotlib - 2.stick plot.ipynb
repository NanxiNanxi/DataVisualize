```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
matplotlib.use('nbagg')
%matplotlib inline
```

```python
np.random.seed(0)
x = np.arange(5)
y= np.random.randn(5)

fig,axes = plt.subplots(ncols = 2)
v_bars = axes[0].bar(x,y,color = 'r')
h_bars = axes[1].barh(x,y,color='b')
```

```python
np.random.seed(0)
x = np.arange(5)
y= np.random.randint(-5,5,5)

fig,axes = plt.subplots(ncols = 2)
v_bars = axes[0].bar(x,y,color = 'r')
h_bars = axes[1].barh(x,y,color='b')
axes[0].axhline(0,color = 'b',linewidth= 2.0)
axes[1].axvline(0,color = 'r',linewidth = 2.0)
```

```python
fig,ax = plt.subplots()
v_bars = ax.bar(x,y,color = 'lightblue')
for bar, height in zip(v_bars,y):
    if height < 0:
        bar.set(edgecolor = 'darkred',color = 'green',linewidth = 3)
ax.axhline(0,color = 'red',linewidth = 2)
```

```python
x1 = np.array([1,2,3])
x2 = np.array([2,2,3])
bar_labels = ['bar1','bar2','bar3']
fig = plt.figure(figsize = (8,6))
y_pos = np.arange(len(x1))
y_pos = [x for x in y_pos]

plt.barh(y_pos,x1,color='g',alpha = 0.3)
plt.barh(y_pos,-x2,color='b',alpha = 0.3)
plt.xlim(-max(x2)-1,max(x1)+1)
plt.ylim(-1,len(x1)+1)
```

```python
g_data = [1,2,3]
b_data = [3,2,1]
r_data = [2,3,3]
labels = ['g1','g2','g3']

pos = list(range(len(g_data)))
width = 0.2
fig,ax = plt.subplots(figsize = (8,6))
plt.bar(pos,g_data,width,alpha=0.5,color ='g',label = labels[0])
plt.bar([p+width for p in pos],b_data,width,alpha=0.5,color ='b',label = labels[1])
plt.bar([p+width*2 for p in pos],r_data,width,alpha=0.5,color ='r',label = labels[2])


```

```python
data = range(200,225,5)
bar_labels = ['a','b','c','d','e']
fig = plt.figure(figsize = (10,8))
y_pos = np.arange(len(data))
plt.yticks(y_pos,bar_labels,fontsize = 16)
bars = plt.barh(y_pos,data,alpha = 0.5,color = 'g')
plt.vlines(min(data),-1,len(data)+0.5,linestyle = 'dashed')
# each bar's percentage of the lowest value 
for b,d in zip(bars,data):
    plt.text(b.get_width()+b.get_width()*0.05, b.get_y()+b.get_height()/2,'{0:.2%}'.format(d/min(data)))
```

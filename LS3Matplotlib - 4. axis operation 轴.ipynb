```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = range(10)
y = range(10)

fig = plt.gca()
plt.plot(x,y)

# set visibility of x and y axis
fig.axes.get_xaxis().set_visible(False)
fig.axes.get_yaxis().set_visible(False)
```

```python
import math
x = np.random.normal(loc=0.0, scale = 1.0, size = 300)
width = 0.5
bins = np.arange(math.floor(x.min())-width, math.ceil(x.max())+width,width)

ax = plt.subplot(111)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tick_params(bottom='off',top='off',left='off')
plt.grid()
plt.hist(x,alpha = 0.5, bins= bins)
```

```python
x = range(10)
y = range(10)

#变量与str的结合
labels = ['label %d'%i for i in range(10)]
fig,ax = plt.subplots()

plt.plot(x,y)

ax.set_xticklabels(labels, rotation = 45, horizontalalignment = 'right')
```

```python
x = np.arange(10)
for i in range(1,4):
    plt.plot(x,i*x**2,label = 'Group %d' %i)
plt.legend(loc = 'best') #'upper left'
```

```python
fig = plt.figure()
ax = plt.subplot(111)
x = np.arange(10)

for i in range(1,4):
    plt.plot(x,i*x**2,label = 'Group %d'%i)
   #bbox_to_anchor设置注解box所处位置, ncol设置一行有几个group
ax.legend(loc = 'upper center',bbox_to_anchor = (0.5,1.15), ncol = 3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25])
plt.xlabel('xlabel',fontsize = 16)
plt.ylabel('ylabel',fontsize = 16)
```

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],'.-')
plt.xlabel('xlabel',fontsize = 16)
plt.ylabel('ylabel',fontsize = 16)
```

字符|类型 | 字符|类型
---|--- | --- | ---
`  '-'	`| 实线 | `'--'`|	虚线
`'-.'`|	虚点线 | `':'`|	点线
`'.'`|	点 | `','`| 像素点
`'o'`	|圆点 | `'v'`|	下三角点
`'^'`|	上三角点 | `'<'`|	左三角点
`'>'`|	右三角点 | `'1'`|	下三叉点
`'2'`|	上三叉点 | `'3'`|	左三叉点
`'4'`|	右三叉点 | `'s'`|	正方点
`'p'`	| 五角点 | `'*'`|	星形点
`'h'`|	六边形点1 | `'H'`|	六边形点2 
`'+'`|	加号点 | `'x'`|	乘号点
`'D'`|	实心菱形点 | `'d'`|	瘦菱形点 
`'_'`|	横线点 | |

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],'.-',color='r')
plt.xlabel('xlabel',fontsize = 16)
plt.ylabel('ylabel',fontsize = 16)
```

```python
array1 = np.arange(0,10,0.5)
plt.plot(array1,array1,'r-')
plt.plot(array1,array1**2,'bs')
plt.plot(array1,array1**3,'go')
```

```python
array1 = np.arange(0,10,0.5)
plt.plot(array1,array1,'r-',
         array1,array1**2,'bs',
         array1,array1**3,'go')
```

```python
x = np.linspace(-10,10)
y = np.sin(x)
plt.plot(x,y)
```

```python
plt.plot(x,y,color = 'b',linestyle=':',marker = 'o',markerfacecolor = 'r',markersize = 10)
```

```python
line = plt.plot(x,y)
plt.setp(line,color = 'r',linewidth = 2.0, alpha = 0.5)
```

### child 子图

#### 211 referes to 2 行 1列, 最后一个数字表示是两个图的第几个

```python
plt.subplot(211)
plt.plot(x,y,color = 'r')

plt.subplot(212)
plt.plot(x,y**2,color = 'b')
```

```python
plt.subplot(121)
plt.plot(x,y,color = 'r')

plt.subplot(122)
plt.plot(x,y**2,color = 'b')
```

```python
plt.subplot(321)
plt.plot(x,y,color = 'r')

plt.subplot(324)
plt.plot(x,y**2,color = 'b')
```

### 为图加注释

```python
plt.plot(x,y,color = 'b',linestyle=':',marker = 'o',markerfacecolor = 'r',markersize = 10)
plt.xlabel('x:---')
plt.ylabel('y:---')
plt.title('test:---')
plt.text(0,0,'text for text')
plt.grid(True)
plt.annotate('test for annotation',xy = (-5,0),xytext = (-2,0.3),arrowprops = dict(facecolor = 'pink',shrink = 0.05,headwidth=10,headlength = 20))
```
